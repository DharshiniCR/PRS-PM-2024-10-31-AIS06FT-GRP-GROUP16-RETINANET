# Prepare Portfolios using the Rules



```python
import numpy as np
import pandas as pd

# Function to create portfolios
def create_portfolios(normalized_data_with_ticker, clusters, portfolio_type_weights='EW'):
    portfolios = {}

    # Track all tickers to include in portfolios
    all_tickers = normalized_data_with_ticker['Ticker'].unique()

    # Create a default portfolio for stocks not in any cluster
    default_portfolio = {}

    # Iterate over each cluster
    for cluster_id, tickers in clusters.items():
        # Extract the stock tickers in the current cluster
        cluster_data = normalized_data_with_ticker[normalized_data_with_ticker['Ticker'].isin(tickers)]
        cluster_data_n = cluster_data.drop(columns=['Ticker'])
        num_ticker = len(tickers)

        # Create portfolios based on the selected weight type
        if portfolio_type_weights == 'EW':
            ew_weights = np.ones(num_ticker) / num_ticker
            portfolio_name = f'EW_Cluster_{cluster_id}'
            portfolios[portfolio_name] = dict(zip(tickers, ew_weights))

        elif portfolio_type_weights == 'MCW':
            market_caps = np.random.uniform(100, 500, num_ticker)  # Simulated market caps
            mcw_weights = market_caps / np.sum(market_caps)
            portfolio_name = f'MCW_Cluster_{cluster_id}'
            portfolios[portfolio_name] = dict(zip(tickers, mcw_weights))

        elif portfolio_type_weights == 'GMV':
            returns = cluster_data_n.pct_change().dropna()
            cov_matrix = returns.cov()
            inv_cov_matrix = np.linalg.inv(cov_matrix)
            ones = np.ones(cov_matrix.shape[0])
            gmv_weights = np.dot(inv_cov_matrix, ones) / np.dot(ones.T, np.dot(inv_cov_matrix, ones))
            portfolio_name = f'GMV_Cluster_{cluster_id}'
            portfolios[portfolio_name] = dict(zip(tickers, gmv_weights))

        elif portfolio_type_weights == 'MSR':
            cluster_n = cluster_data_n.pct_change().dropna()
            cluster_exp_returns = cluster_n.mean() * 252  # Annualized returns
            cov_matrix = cluster_n.cov()
            inv_cov_matrix = np.linalg.inv(cov_matrix)
            ones = np.ones((len(cluster_exp_returns), 1))
            numer = np.dot(inv_cov_matrix, cluster_exp_returns)
            denom = np.dot(ones.T, numer)
            msr_weights = numer / denom
            portfolio_name = f'MSR_Cluster_{cluster_id}'
            portfolios[portfolio_name] = dict(zip(tickers, msr_weights.flatten()))

    # Create a default portfolio for stocks not in any cluster
    for ticker in all_tickers:
        if ticker not in [tick for cluster in clusters.values() for tick in cluster]:
            default_portfolio[ticker] = 1 / len(all_tickers)  # Assign equal weight to all non-clustered stocks

    # Add the default portfolio to the portfolios dictionary
    if default_portfolio:
        portfolios['Default_Portfolio'] = default_portfolio

    return portfolios
```

# Simulated Annealing Optimization

High Return Optimization

```python
import numpy as np

# Objective function for Simulated Annealing: return and volatility calculation
def portfolio_performance(weights, returns):
    # Calculate portfolio return
    portfolio_return = np.sum(weights * returns)

    # Calculate portfolio volatility (assuming equal covariance for simplicity)
    portfolio_volatility = np.std(weights * returns)

    return portfolio_return, portfolio_volatility

# Simulated Annealing function
def simulated_annealing(returns, num_assets, initial_weights, iterations=5000, temperature=1000):
    current_weights = np.copy(initial_weights)
    current_return, current_volatility = portfolio_performance(current_weights, returns)

    best_weights = np.copy(current_weights)
    best_return, best_volatility = current_return, current_volatility

    for i in range(iterations):
        # Generate new candidate weights by making small random changes
        candidate_weights = current_weights + np.random.normal(0, 0.1, size=num_assets)
        candidate_weights = np.clip(candidate_weights, 0, 1)  # Ensure weights are between 0 and 1
        candidate_weights /= np.sum(candidate_weights)  # Ensure weights sum to 1

        candidate_return, candidate_volatility = portfolio_performance(candidate_weights, returns)

        # Accept new candidate if it's better or with some probability if worse (simulated annealing logic)
        delta = candidate_return - current_return
        if delta > 0 or np.exp(delta / temperature) > np.random.rand():
            current_weights = candidate_weights
            current_return, current_volatility = candidate_return, candidate_volatility

            # Update best solution if the candidate is better
            if candidate_return > best_return:
                best_weights = candidate_weights
                best_return, best_volatility = candidate_return, candidate_volatility

        # Gradually cool down the temperature
        temperature *= 0.99

    return best_weights, best_return, best_volatility
```

```python
# Objective function for Low Risk optimization: prioritize minimizing volatility
def portfolio_performance_low_risk(weights, returns):
    # Calculate portfolio return
    portfolio_return = np.sum(weights * returns)

    # Calculate portfolio volatility (assuming equal covariance for simplicity)
    portfolio_volatility = np.std(weights * returns)

    return portfolio_return, portfolio_volatility

# Simulated Annealing function for Low Risk
def simulated_annealing_low_risk(returns, num_assets, initial_weights, iterations=5000, temperature=1000):
    current_weights = np.copy(initial_weights)
    current_return, current_volatility = portfolio_performance_low_risk(current_weights, returns)

    best_weights = np.copy(current_weights)
    best_return, best_volatility = current_return, current_volatility

    for i in range(iterations):
        # Generate new candidate weights by making small random changes
        candidate_weights = current_weights + np.random.normal(0, 0.1, size=num_assets)
        candidate_weights = np.clip(candidate_weights, 0, 1)  # Ensure weights are between 0 and 1
        candidate_weights /= np.sum(candidate_weights)  # Ensure weights sum to 1

        candidate_return, candidate_volatility = portfolio_performance_low_risk(candidate_weights, returns)

        # Accept new candidate if it's better or with some probability if worse (simulated annealing logic)
        delta = current_volatility - candidate_volatility  # Prioritize volatility reduction
        if delta > 0 or np.exp(delta / temperature) > np.random.rand():
            current_weights = candidate_weights
            current_return, current_volatility = candidate_return, candidate_volatility

            # Update best solution if the candidate is better (lower volatility)
            if candidate_volatility < best_volatility:
                best_weights = candidate_weights
                best_return, best_volatility = candidate_return, candidate_volatility

        # Gradually cool down the temperature
        temperature *= 0.99

    return best_weights, best_return, best_volatility
```
